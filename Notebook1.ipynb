```python
family_level_insights = df.groupby("Family ID").agg(
    
    Total_spending = ("Amount" , "sum"),
    Total_savings = ("Savings" , 'sum'),
    Monthly_expanses = ('Monthly Expenses' , 'mean'),
    Loan_payment = ('Loan Payments', 'sum'),
    average_spending = ('Amount', 'mean'),
    credit_card_spending = ('Credit Card Spending', 'sum'),
    Category_spending=('Category', lambda x: x.value_counts(normalize=True).to_dict())
    
).reset_index()
```

```python
family_level_insights['Saving_ratio_spending'] = (df.groupby('Family ID').apply(lambda x :(x['Savings'] / x['Income'].sum()))).reset_index(drop=True)
```

```python
family_level_insights = family_level_insights.merge(income, on='Family ID', how='left')
family_level_insights.head()
```
